```python
def extract_format(dicts,super=super):
    code=""
    for i,j in dicts.items():
        count=1
        if "[" in i and "]" in i:
            count=int(i[i.index("[")+1:i.index("]")])
        #print(j)
        if len(j)>1:
            j=extract_format(super[j])
        code=code+j*count
    return code
```

```python
def event_format(bits,offset_bits='<H4BQfI2B',bits_selected='BBBB',event_formattable=event_formattable,super=super):
    offset,length=struct.calcsize(offset_bits),struct.calcsize(offset_bits+bits_selected)
    string_code=struct.unpack(bits_selected,bits[offset:offset+length])
    if string_code in event_formattable:
        return offset_bits+bits_selected+extract_format(super["EventDataDetails"][string_code])
    return offset_bits+bits_selected
```
